## Filling Missing Dates Using Data from the two Files

```python
import pandas as pd
```

```python
df1 = pd.read_csv("\\Users\\laura\\Desktop\\Progetto finale\\Videogames\\metacritic_data.csv")
```

```python
df2 = pd.read_csv("\\Users\\laura\\Desktop\\Progetto finale\\Videogames\\vgchartz_sales.csv")
```

```python
df2 = df2.rename(columns={'videogame': 'Game_name'})
```

```python
df2 = df2.rename(columns={'release': 'Game_release_date'})
```

```python
df1['Game_name'] = df1['Game_name'].str.title()
```

```python
df2['Game_name'] = df2['Game_name'].str.title()
```

```python
# Converting strings to datatypes
df1['Game_release_date'] = pd.to_datetime(df1['Game_release_date'], format='%d/%m/%Y', errors='coerce', dayfirst=True)
df2['Game_release_date'] = pd.to_datetime(df2['Game_release_date'], format='%d/%m/%Y', errors='coerce', dayfirst=True)
```

```python
release_date_dict = df2.set_index('Game_name')['Game_release_date'].to_dict()
```

Filling empty dates in df1

```python
df1['Game_release_date'] = df1['Game_release_date'].fillna(df1['Game_name'].map(release_date_dict))
```

```python
df1['Game_release_date'].isna().sum()
```

```python
no_date = df1[df1['Game_release_date'].isna()]
no_date
```

```python
oldest_dates = df2.groupby('Game_name')['Game_release_date'].min().dropna().to_dict()
```

```python
def fill_missing_dates(row, date_dict):
    if pd.isna(row['Game_release_date']):
        return date_dict.get(row['Game_name'], row['Game_release_date'])
    return row['Game_release_date']
```

```python
df1['Game_release_date'] = df1.apply(fill_missing_dates, axis=1, date_dict=oldest_dates)
```

```python
df1['Game_release_date'].isna().sum()
```

```python
no_date = df1[df1['Game_release_date'].isna()]
no_date
```

Filling missing dates df2

```python
df2
```

```python
release_date_dict_df1 = df1.set_index('Game_name')['Game_release_date'].to_dict()
```

```python
df2['Game_release_date'] = df2['Game_release_date'].fillna(df2['Game_name'].map(release_date_dict_df1))
```

```python
df2['Game_release_date'].isna().sum()
```

```python
oldest_dates = df2.groupby('Game_name')['Game_release_date'].min()

# Riempire i valori mancanti nella colonna Game_release_date con le date piÃ¹ vecchie
df2['Game_release_date'] = df2['Game_release_date'].fillna(df2['Game_name'].map(oldest_dates))
```

```python
df2['Game_release_date'].isna().sum()
```

```python
no_date2 = df2[df2['Game_release_date'].isna()]
no_date2
```
