## Data cleaning Vgchartz

```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
df_vg = pd.read_csv("\\Users\\laura\\Desktop\\Progetto finale\\Videogames\\vgchartz_sales_uncleaned.csv")
```

```python
df_vg
```

```python
df_vg.dropna(subset=['videogame'])
```

```python
df_vg['release'].isna().sum()
```

```python
no_date = df_vg[df_vg['release'].isna()]
no_date
```

```python
def convert_date(date_str):
    if date_str.strip(): 
        date_str = date_str.replace("st", "").replace("nd", "").replace("rd", "").replace("th", "")
        date = datetime.strptime(date_str, "%d %b %y")
        formatted_date = date.strftime("%d/%m/%Y")
        return formatted_date
    else:
        return ""
```

```python
df_vg['release'] = df_vg['release'].fillna("")
```

```python
df_vg['release'] = df_vg['release'].apply(convert_date)
```

```python
df_vg
```

```python
df_vg = df_vg.drop_duplicates()
```

```python
df_vg
```

```python
def convert_to_millions_as_int(number):
    if pd.notnull(number): 
        return int(number * 1_000_000)
    else:
        return None
```

```python
df_vg['sales_na'] = df_vg['sales_na'].apply(convert_to_millions_as_int)
```

```python
df_vg['sales_eu'] = df_vg['sales_eu'].apply(convert_to_millions_as_int)
```

```python
df_vg['sales_jp'] = df_vg['sales_jp'].apply(convert_to_millions_as_int)
```

```python
df_vg['sales_otras'] = df_vg['sales_otras'].apply(convert_to_millions_as_int)
```

```python
df_vg['sales_tot'] = df_vg['sales_tot'].apply(convert_to_millions_as_int)
```

```python
df_vg.loc[df_vg['release'] == '', 'release'] = np.nan
```

```python
df_vg
```

```python
df_vg.to_csv('vgchartz_sales.csv', index=False)
```
