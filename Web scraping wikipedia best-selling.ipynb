```python
import pandas as pd
from bs4 import BeautifulSoup
import requests, time, datetime, re, random
```

```python
wiki_url = "https://en.wikipedia.org/wiki/List_of_best-selling_video_games"
headers = {'User-Agent' : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36"}
response = requests.get(wiki_url, headers=headers)
response.status_code
```

```python
# parse data from the html into a beautifulsoup object
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
# find the table of vg sales
html_table = soup.find('table',{'class':'wikitable'})
html_table
```

```python
# convert the wikipedia table into a pandas dataframe
vg_sales_df = pd.read_html(str(html_table))[0]
vg_sales_df
```

```python
vg_sales_df['Sales'] = vg_sales_df['Sales'].astype(int)
```

```python
vg_sales_df
```

```python
first_date = vg_sales_df['Initial release date'][0]
```

```python
date_cleaned = first_date.split('[')[0].strip()
```

```python
date_cleaned
```

```python
def remove_suffix(c):
    if '[c]' in c:
        return c.split('[c]')[0].strip()
    return c
```

```python
vg_sales_df['Initial release date'] = vg_sales_df['Initial release date'].apply(remove_suffix)
```

```python
vg_sales_df
```

```python
# Initial release date

def convert_date(date_str):
    date_pattern = r'(\w+\s\d{1,2},\s\d{4})'
    match = re.search(date_pattern, date_str)
    if match:
            extracted_date = match.group(1)
            datetime_obj = datetime.datetime.strptime(extracted_date, '%B %d, %Y')
            return datetime_obj.strftime('%m/%d/%Y')
    else:
          return None
```

```python
vg_sales_df['Initial release date'] = vg_sales_df['Initial release date'].apply(convert_date)
vg_sales_df
```

```python
vg_sales_df.drop(columns=['Ref.'], inplace=True)
```

```python
vg_sales_df
```

```python
with pd.ExcelWriter('List_of_best-selling_video_games.xlsx') as writer:
    vg_sales_df.to_excel(writer, sheet_name='Best-selling')
```
