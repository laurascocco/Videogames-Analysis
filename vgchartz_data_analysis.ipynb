```python
platform_title = df_vg.platform.value_counts()

plt.subplots(figsize=(10,15))
ax = sns.barplot(x=platform_title , y=platform_title .index, palette='cubehelix')
ax.set_title('Most Number Titles per Platform', color='red', alpha=0.5, size=25)
ax.set_xlabel('Total Title', color='green', alpha=0.5, size=30)
ax.set_ylabel('Platform', color='green', alpha=0.5, size=30)
```

```python
games_by_genre = df_vg.groupby('genre').agg(total_games=('genre', 'count')).sort_values('total_games')
plt.subplots(figsize=(28,10))
ax = sns.pointplot(x=games_by_genre.index, y=games_by_genre.total_games)
ax.set_title('Total Number of Games by Genre', color='blue', size=25, alpha=0.5)
ax.set_xlabel('Genre', color='green', size=25, alpha=0.5)
ax.set_ylabel('Total Number of Games', color='green', size=25, alpha=0.5)
```

```python
global_sales_by_genre = df_vg.groupby('genre').agg({'sales_tot':np.sum}).sort_values('sales_tot')
plt.subplots(figsize=(20,8))
ax = sns.barplot(x=global_sales_by_genre.index, y=global_sales_by_genre.sales_tot)
ax.set_title('Global Sales of Games by Genre', color='blue', size=25, alpha=0.5)
ax.set_xlabel('Genre', color='green', size=25, alpha=0.5)
ax.set_ylabel('Global Sales', color='green', size=25, alpha=0.5)
```

```python
global_sales_publisher = df_vg.pivot_table(index=['editor'], values=['sales_tot','sales_na','sales_eu','sales_jp', 'sales_otras'], 
                                           aggfunc=np.sum).sort_values(['sales_tot'], ascending=False)
global_sales_publisher = global_sales_publisher[['sales_tot','sales_na','sales_eu','sales_jp', 'sales_otras']]
ax = global_sales_publisher.iloc[0:10,:].plot(kind='bar', stacked=True, grid=False)
ax.set_title('Top 10 Publishers by Global Sales', size=25, color='blue', alpha=0.5)
ax.set_xlabel('Publisher', size=25, color='green', alpha=0.5)
ax.set_ylabel('Sales', size=25, color='green', alpha=0.5)
```

```python
global_sales_developer = df_vg.pivot_table(index=['developer'], values=['sales_tot','sales_na','sales_eu','sales_jp', 'sales_otras'], 
                                           aggfunc=np.sum).sort_values(['sales_tot'], ascending=False)
global_sales_developer = global_sales_developer[['sales_tot','sales_na','sales_eu','sales_jp', 'sales_otras']]
ax = global_sales_developer.iloc[0:10,:].plot(kind='bar', stacked=True, grid=False)
ax.set_title('Top 10 Developers by Global Sales', size=25, color='blue', alpha=0.5)
ax.set_xlabel('Developer', size=25, color='green', alpha=0.5)
ax.set_ylabel('Sales', size=25, color='green', alpha=0.5)
plt.show()
```

```python
df_vg['release'] = pd.to_datetime(df_vg['release'], format='%d/%m/%Y', errors='coerce')

df_vg = df_vg.dropna(subset=['release'])

df_vg['Year_of_Release'] = df_vg['release'].dt.year.astype(int)
```

```python
titles_by_year = df_vg.groupby(['Year_of_Release']).agg({'videogame':lambda x : len(x) if (len(x) > 1) else None}).dropna()
plt.subplots(figsize=(12,7))
ax = sns.pointplot(x=titles_by_year.index, y=titles_by_year.videogame)
g = ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
ax.set_title('Number of Titles by Year of Release', size=25, color='blue', alpha=0.5)
ax.set_xlabel('Year of Release', size=25, color='red', alpha=0.5)
ax.set_ylabel('Number of Titles', size=25, color='red', alpha=0.5)
```

```python
platform_sales = df_vg.groupby("platform")[["sales_na", "sales_eu", "sales_jp", "sales_otras"]].sum()

# Crea il grafico a barre orizzontali impilate
platform_sales.plot(kind='barh', stacked=True, figsize=(8, 12))
plt.xlabel('Total Sales')
plt.ylabel('Platform')
plt.title('Total Sales by Platform and Region')
plt.legend(title='Region')
plt.grid(axis='x')
plt.show()
```

```python
platform_sales = df_vg.groupby("platform")[["sales_na", "sales_eu", "sales_jp", "sales_otras"]].sum()

# Calcola le vendite totali per ogni piattaforma
platform_sales['total_sales'] = platform_sales.sum(axis=1)

# Ordina il DataFrame in base alle vendite totali in ordine decrescente
platform_sales_sorted = platform_sales.sort_values(by='total_sales', ascending=True)

# Crea il grafico a barre orizzontali impilate con le barre in ordine dal più alto al più basso
platform_sales_sorted.drop(columns='total_sales').plot(kind='barh', stacked=True, figsize=(8, 12), legend=True)
plt.xlabel('Total Sales')
plt.ylabel('Platform')
plt.title('Total Sales by Platform and Region (Sorted)')
plt.legend(title='Region')
plt.grid(axis='x')
plt.show()
```

```python
editor_sales = df_vg.groupby("editor")["sales_tot"].sum().reset_index().sort_values(by="sales_tot", ascending=False)

# Raggruppa il DataFrame per sviluppatore e calcola le vendite totali
developer_sales = df_vg.groupby("developer")["sales_tot"].sum().reset_index().sort_values(by="sales_tot", ascending=False)

# Crea due grafici a barre separati per editor e sviluppatore
plt.figure(figsize=(12, 8))

plt.subplot(2, 1, 1)
plt.bar(editor_sales["editor"][:10], editor_sales["sales_tot"][:10], color='blue')
plt.xlabel('Editor')
plt.ylabel('Total Sales')
plt.title('Top 10 Editors by Total Sales')
plt.xticks(rotation=45, ha='right')

plt.subplot(2, 1, 2)
plt.bar(developer_sales["developer"][:10], developer_sales["sales_tot"][:10], color='green')
plt.xlabel('Developer')
plt.ylabel('Total Sales')
plt.title('Top 10 Developers by Total Sales')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
df_vg['release'] = pd.to_datetime(df_vg['release'])

# Raggruppa il DataFrame per anno e calcola le vendite totali
yearly_sales = df_vg.groupby(df_vg['release'].dt.year)['sales_tot'].sum()

# Crea il grafico a linea
plt.figure(figsize=(10, 6))
plt.plot(yearly_sales.index, yearly_sales.values, marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.title('Total Sales Over Time')
plt.grid(True)
plt.show()
```

```python
df_vg['release'] = pd.to_datetime(df_vg['release'])

# Raggruppa il DataFrame per anno e calcola le vendite totali
yearly_sales = df_vg.groupby(df_vg['release'].dt.year)['sales_tot'].sum()

# Calcola la media mobile delle vendite con una finestra temporale di 3 anni
rolling_avg_sales = yearly_sales.rolling(window=3).mean()

# Crea il grafico a linea
plt.figure(figsize=(10, 6))
plt.plot(yearly_sales.index, yearly_sales.values, marker='o', linestyle='-', label='Total Sales')
plt.plot(rolling_avg_sales.index, rolling_avg_sales.values, linestyle='--', color='red', label='3-Year Rolling Average')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.title('Total Sales Over Time with Rolling Average')
plt.legend()
plt.grid(True)
plt.show()
```
