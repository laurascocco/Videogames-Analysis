```python
total_sales_editor = df_vg.groupby("editor")["sales_tot"].sum().reset_index()
total_sales_editor["sales_editor"] = total_sales_editor["sales_tot"]
total_sales_editor = total_sales_editor.sort_values(by="sales_tot", ascending=False)
```

```python
pd.options.display.float_format = '{:,.2f}'.format
total_sales_editor[["editor", "sales_editor"]]
```

```python
colors = plt.cm.viridis(np.linspace(0, 1, len(top_20_editor)))

plt.figure(figsize=(10, 6))
plt.bar(total_sales_editor["editor"].head(20), total_sales_editor["sales_editor"].head(20), color=colors)
plt.title('Top 20 Editor Total Sales', fontsize=16)
plt.xlabel('Editor')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```

```python
videogame_platform = df_vg.groupby("Game_name")["platform"].count().reset_index()
videogame_platform["tot_platform"] = videogame_platform["platform"]
videogame_platform = videogame_platform.sort_values(by="platform", ascending=False)
```

```python
global_sales_by_genre = df_vg.groupby('genre').agg({'sales_tot':np.sum}).sort_values('sales_tot')

colors = plt.cm.viridis(np.linspace(0, 1, len(global_sales_by_genre)))[::-1]

plt.figure(figsize=(10, 8))
plt.barh(global_sales_by_genre.index[-20:], global_sales_by_genre["sales_tot"].tail(20), color=colors)
plt.title('Top 20 Genre Total Sales', fontsize=16)
plt.xlabel('Total Sales', fontsize=12)
plt.ylabel('Genre', fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Convert the "Game_release_date" column to datetime to extract the year.
df_vg['Year_of_Release'] = pd.to_datetime(df_vg['Game_release_date']).dt.year

titles_by_year = df_vg.groupby('Year_of_Release').size()

plt.figure(figsize=(12, 7))
plt.plot(titles_by_year.index, titles_by_year.values, marker='o', color='purple')
plt.title('Number of Titles by Year of Release', fontsize=16)
plt.xlabel('Year of Release', fontsize=12)
plt.ylabel('Number of Titles', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
platform_sales = df_vg.groupby("platform")[["sales_na", "sales_eu", "sales_jp", "sales_otras"]].sum()

platform_sales['total_sales'] = platform_sales.sum(axis=1)

platform_sales_sorted = platform_sales.sort_values(by='total_sales', ascending=False).head(20)

platform_sales_sorted = platform_sales_sorted.sort_values(by='total_sales', ascending=True) 
platform_sales_sorted.drop(columns='total_sales').plot(kind='barh', stacked=True, figsize=(10, 8), legend=True)
plt.xlabel('Total Sales')
plt.ylabel('Platform')
plt.title('Top 20 Platforms by Total Sales and Region')
plt.legend(title='Region')
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```

```python
sales_by_year = df_vg.groupby('Year_of_Release')['sales_tot'].sum()

plt.figure(figsize=(12, 7))
plt.plot(sales_by_year.index, sales_by_year.values, marker='o', color='purple')
plt.title('Total Sales by Year of Release', fontsize=16)
plt.xlabel('Year of Release', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.grid(True)
plt.show()
```
