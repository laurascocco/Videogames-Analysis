```python
reviews_by_genre = df_vg.groupby('Game_genre').agg({'Critic_num_reviews': 'sum', 'User_num_reviews': 'sum'})

# Ordina il DataFrame in base al numero totale di recensioni
reviews_by_genre = reviews_by_genre.sum(axis=1).sort_values(ascending=False)

# Plot
plt.subplots(figsize=(10, 25))
ax = sns.barplot(x=reviews_by_genre, y=reviews_by_genre.index, palette='cubehelix')
ax.set_title('Number of Reviews by Genre', color='red', alpha=0.5, size=25)
ax.set_xlabel('Total Reviews', color='green', alpha=0.5, size=30)
ax.set_ylabel('Genre', color='green', alpha=0.5, size=30)
```

```python
df_vg['Release_Year'] = pd.to_datetime(df_vg['Game_release_date']).dt.year

# Raggruppa per anno di uscita e calcola il numero totale di recensioni
reviews_by_year = df_vg.groupby('Release_Year').agg({'Critic_num_reviews': 'sum', 'User_num_reviews': 'sum'})

# Plot
plt.figure(figsize=(12, 6))
plt.scatter(reviews_by_year.index, reviews_by_year.sum(axis=1), color='blue')
plt.title('Number of Reviews by Release Year')
plt.xlabel('Release Year')
plt.ylabel('Total Reviews')
plt.grid(True)
plt.tight_layout()
```

```python
df_vg['Total_reviews'] = df_vg[['Critic_num_reviews', 'User_num_reviews']].sum(axis=1)

# Converti le colonne delle recensioni dei critici, degli utenti e delle recensioni totali in interi
df_vg[['Critic_num_reviews', 'User_num_reviews', 'Total_reviews']] = df_vg[['Critic_num_reviews', 'User_num_reviews', 'Total_reviews']].fillna(0).astype(int)

# Seleziona i top 10 giochi con il numero piÃ¹ alto di recensioni
top_10_games = df_vg.nlargest(10, 'Total_reviews')[['Game_name', 'Game_release_date', 'Critic_num_reviews', 'Critic_score_0_100', 'User_num_reviews', 'User_score_0_10', 'Total_reviews']]

# Visualizza i top 10 giochi con numeri interi
top_10_games
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df_vg['Critic_score_0_100'].dropna(), bins=20, color='blue', alpha=0.7)
plt.xlabel('Critic Score (0-100)')
plt.ylabel('Frequency')
plt.title('Distribution of Critic Review Scores')
plt.grid(True)
plt.show()
```

```python
df_vg['Game_release_date'] = pd.to_datetime(df_vg['Game_release_date'])

# Extract year from 'Game_release_date'
df_vg['Year'] = df_vg['Game_release_date'].dt.year

# Group by year and calculate average critic and user scores
avg_scores = df_vg.groupby('Year').agg({'Critic_score_0_100': 'mean', 'User_score_0_10': lambda x: x.mean() * 10})

# Plot the average scores over the years
plt.figure(figsize=(10, 6))
plt.plot(avg_scores.index, avg_scores['Critic_score_0_100'], marker='o', color='blue', label='Average Critic Score')
plt.plot(avg_scores.index, avg_scores['User_score_0_10'], marker='o', color='green', label='Average User Score (scaled)')
plt.xlabel('Year')
plt.ylabel('Average Score')
plt.title('Average Critic and User Scores Over the Years')
plt.grid(True)
plt.legend()
plt.show()
```
