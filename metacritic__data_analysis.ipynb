```python
total_reviews_critic_distribution = (df_vg.groupby("Game_distribution")["Critic_num_reviews"]
                                      .sum()
                                      .reset_index(name="reviews_distribution")
                                      .sort_values(by="reviews_distribution", ascending=False))
```

```python
total_reviews_critic_distribution
```

```python
top_20_reviews = total_reviews_critic_distribution.head(20)

colors = plt.cm.viridis(np.linspace(0, 1, len(top_20_reviews)))

plt.figure(figsize=(10, 6))
plt.bar(top_20_reviews["Game_distribution"], 
        top_20_reviews["reviews_distribution"], 
        color=colors)
plt.title('Top 20 Total Critic Reviews by Distribution', fontsize=16)
plt.xlabel('Game Distribution', fontsize=14)
plt.ylabel('Total Critic Reviews', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```

```python
total_reviews_user_distribution = (df_vg.groupby("Game_distribution")["User_num_reviews"]
                                      .sum()
                                      .reset_index(name="reviews_distribution")
                                      .sort_values(by="reviews_distribution", ascending=False))
```

```python
total_reviews_user_distribution
```

```python
top_20_reviews = total_reviews_user_distribution.head(20)

colors = plt.cm.viridis(np.linspace(0, 1, len(top_20_reviews)))

plt.figure(figsize=(10, 6))
plt.bar(top_20_reviews["Game_distribution"], 
        top_20_reviews["reviews_distribution"], 
        color=colors)
plt.title('Top 20 Total User Reviews by Distribution', fontsize=16)
plt.xlabel('Game Distribution', fontsize=14)
plt.ylabel('Total User Reviews', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```

```python
reviews_by_platform_top20 = (df_vg.groupby('Game_platform')
                             .agg({'Critic_score_0_100': 'mean', 'User_score_0_10': lambda x: x.mean() * 10})
                             .head(20))

colors = ['purple', 'gold'] * (len(reviews_by_platform_top20) // 2)

plt.figure(figsize=(10, 6))
reviews_by_platform_top20[['Critic_score_0_100', 'User_score_0_10']].plot(kind='bar', color=colors, width=0.8)
plt.title('Top 20 Average Critic and User Scores by Platform', fontsize=16)
plt.xlabel('Platform', fontsize=14)
plt.ylabel('Average Score', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.legend(['Critic Score', 'User Score (x10)'], fontsize=12)
plt.tight_layout()
plt.show()
```

```python
reviews_by_platform = df_vg.groupby('Game_platform').agg({'Critic_num_reviews': 'sum', 'User_num_reviews': 'sum'})

plt.figure(figsize=(10, 6))
reviews_by_platform.plot(kind='bar', color = ['purple', 'gold'], stacked=True)
plt.title('Number of Reviews by Platform')
plt.xlabel('Platform')
plt.ylabel('Number of Reviews')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
```

```python
top_20_reviews_by_genre = df_vg.groupby('Game_genre').agg({'Critic_num_reviews': 'sum', 'User_num_reviews': 'sum'}).sum(axis=1).nlargest(20)

colors = plt.cm.viridis(np.linspace(0, 1, len(top_20_reviews_by_genre)))

plt.figure(figsize=(10, 6))
plt.bar(top_20_reviews_by_genre.index, 
        top_20_reviews_by_genre.values, 
        color=colors)
plt.title('Top 20 Genres by Number of Reviews', fontsize=16, color='red', alpha=0.5)
plt.xlabel('Genre', fontsize=14)
plt.ylabel('Total Reviews', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```

```python
df_vg['Release_Year'] = pd.to_datetime(df_vg['Game_release_date']).dt.year

reviews_by_year = df_vg.groupby('Release_Year').agg({'Critic_num_reviews': 'sum', 'User_num_reviews': 'sum'})

plt.figure(figsize=(12, 6))
plt.scatter(reviews_by_year.index, reviews_by_year.sum(axis=1), color='blue')
plt.title('Number of Reviews by Release Year', fontsize=16)
plt.xlabel('Release Year', fontsize=12)
plt.ylabel('Total Reviews', fontsize=12)
plt.grid(True)
plt.tight_layout()
```

What happend in 2020

```python
reviews_2020 = df_vg[df_vg['Release_Year'] == 2020][['Game_name', 'Critic_num_reviews', 'User_num_reviews']]

reviews_2020[['Critic_num_reviews', 'User_num_reviews']] = reviews_2020[['Critic_num_reviews', 'User_num_reviews']].fillna(pd.NA).astype('Int64')

reviews_2020 = reviews_2020.sort_values(by=['User_num_reviews', 'Critic_num_reviews'], ascending=False)

reviews_2020
```

```python
top_10_reviews_2020 = reviews_2020.head(10)

colors = plt.cm.viridis(np.linspace(0, 1, len(top_10_reviews_2020)))

plt.figure(figsize=(10, 6))
plt.bar(top_10_reviews_2020["Game_name"], 
        top_10_reviews_2020["User_num_reviews"], 
        color=colors)
plt.title('Top 10 User Reviews for 2020', fontsize=16)
plt.xlabel('Game Name', fontsize=14)
plt.ylabel('Total User Reviews', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```

```python
total_reviews_2020 = int(reviews_2020[['Critic_num_reviews', 'User_num_reviews']].sum().sum())
total_reviews_2020
```

```python
df_vg['Total_reviews'] = df_vg[['Critic_num_reviews', 'User_num_reviews']].fillna(0).sum(axis=1).astype(int)

top_10_games = df_vg.nlargest(10, 'Total_reviews')[['Game_name', 'Game_release_date', 'Critic_num_reviews', 'Critic_score_0_100', 'User_num_reviews', 'User_score_0_10', 'Total_reviews']]

top_10_games
```

```python
top_10_games = df_vg.nlargest(10, 'Total_reviews')
colors = plt.cm.viridis(np.linspace(0, 1, len(top_10_games)))

plt.figure(figsize=(10, 6))
plt.bar(top_10_games["Game_name"], 
        top_10_games["Total_reviews"], 
        color=colors)
plt.title('Top 10 Games by Total Reviews', fontsize=16)
plt.xlabel('Game Name', fontsize=14)
plt.ylabel('Total Reviews', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df_vg['Critic_score_0_100'].dropna(), bins=40, color='purple', alpha=0.7)
plt.xlabel('Critic Score (0-100)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of Critic Review Scores', fontsize=16)
plt.grid(True)
plt.show()
```

```python
user_score_scaled = df_vg['User_score_0_10'].dropna() * 10

plt.figure(figsize=(10, 6))
plt.hist(user_score_scaled, bins=40, color='darkolivegreen', alpha=0.7)
plt.xlabel('User Score (0-100)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of User Review Scores', fontsize=16)
plt.grid(True)
plt.show()
```

```python
avg_critic_score = df_vg.groupby('Release_Year')['Critic_score_0_100'].mean()

avg_user_score = df_vg.groupby('Release_Year')['User_score_0_10'].mean() * 10

plt.figure(figsize=(12, 6))
plt.plot(avg_critic_score.index, avg_critic_score, marker='o', color='blue', label='Average Critic Score')
plt.plot(avg_user_score.index, avg_user_score, marker='o', color='green', label='Average User Score (x10)')
plt.title('Average Critic and User Scores by Release Year', fontsize=16)
plt.xlabel('Release Year', fontsize=12)
plt.ylabel('Average Score', fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
