```python
# Convert date
def convert_date(string):
    if pd.isna(string): 
        return None
    else:
        try:
            #'MMM DD, YYYY'
            date = datetime.strptime(string, '%b %d, %Y')
            return date.strftime('%d/%m/%Y')
        except ValueError:
            try:
                #'MMM YYYY'
                date = datetime.strptime(string, '%b %Y')
                return date.strftime('01/%m/%Y')
            except ValueError:
                try:
                    #'YYYY'
                    date = datetime.strptime(string, '%Y')
                    return date.strftime('01/01/%Y')
                except ValueError:
                    return None
```

```python
df_vg['Game_release_date'] = df_vg['Game_release_date'].apply(convert_date)
```
